Como é possível observar todas as variáveis numéricas apresentadas acima possuem suas disribuições enviesadas. 

Como medida para reduzir o envisamento nos dados, iremos aplicar log em todas as variáveis numéricas que serão utilizadas na predição. Além disso, iremos remover valores NA dos presentes nos dados.

A função abaixo remove os valores NA das variáveis numéricas e substitui pela média dos valores ao qual a coluna pertence.

```python
def remove_NA(data):
    #remove NA das variáveis e substitui pela média
    return data.fillna(data.mean())
```

Para aplicar o log nas variáveis, a fim de diminuir seu enviesamento a função abaixo foi criada.

```python
def set_log_numeric_variables(data):
    numeric_feats = data.dtypes[data.dtypes != "object"].index
    
    #coloca valores das variáveis numéricas em escala de log
    data[numeric_feats] = np.log1p(data[numeric_feats])

    return data
```

Para normalizar as variáveis presentes nos dados comm valores entre 0 a 1, a função de normalização abaixo foi criada.

```python
def data_normalization(data):
    #log transform skewed numeric features:
    numeric_feats = data.dtypes[data.dtypes != "object"].index
    
    #identifica as variáveis categóricas
    cat_feats = data.dtypes[data.dtypes == "object"].index

    #remove variável alvo
    numeric_feats = numeric_feats.drop('votos')

    min_max = preprocessing.MinMaxScaler()
    data_normalized = min_max.fit_transform(data[numeric_feats])
    
    data_normalized = pd.DataFrame(data_normalized, columns=data[numeric_feats].columns)

    #adicionando a coluna votos nos dados já normalizados
    data_normalized = data_normalized.assign(votos = data["votos"].values)
    
    data = pd.concat([data_normalized, data[cat_feats].reset_index()], axis = 1, sort = False)
    
    return data
```

Aqui vamos aplicar todas as funções descritas anteriormente e transformar as variáveis categóricas em DUMMIES variáveis para serem usadas corretamento com os modelos.

```python
def data_proccess(data):
    #1. aplica log nas variáveis numéricas
    data = set_log_numeric_variables(data)
    
    #2. substitui NA pela média da coluna
    data = remove_NA(data)
    
    #3. normaliza variáveis
    data = data_normalization(data)
    
    return data
```

Aplicando normalização nos dados de treino e teste.

```python
train = data_proccess(train)
test = data_proccess(test)
```

```python
all_data = pd.concat([train, test], keys=["train", "test"])
train = pd.get_dummies(all_data).loc["train"]
test = pd.get_dummies(all_data).loc["test"]
```

### Modelagem

Aqui preparamos os dados para serem utilizados nos modelos que serão apresentado nas próximas seções.

```python
#creating matrices for sklearn:
X_train = train.drop('votos', axis=1)
y = train.votos
```

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, LassoCV, LassoLarsCV, Lasso
from sklearn.model_selection import cross_val_score

def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y, scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

### Ridge

```python
model_ridge = Ridge()
```

```python
alphas = [0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75]
cv_ridge = [rmse_cv(Ridge(alpha = alpha)).mean() 
            for alpha in alphas]
```

```python
cv_ridge = pd.Series(cv_ridge, index = alphas)
cv_ridge.plot(title = "Validation")
plt.xlabel("alpha")
plt.ylabel("rmse")
```

```python
cv_ridge.min()
```

Como é observado para regressão utilizando o modelo Ridge o menor RMSE=0.9510488 é apresentado com alpha = 5. Abaixo é apresentado os resíduos com este modelo com alpha = 5.

```python
model_ridge = Ridge(alpha = 5)
model_ridge.fit(X_train, y)

matplotlib.rcParams['figure.figsize'] = (12, 6.0)

preds = pd.DataFrame({"preds":model_ridge.predict(X_train), "true":y})
preds["residuals"] = preds["true"] - preds["preds"]
preds.plot(x = "preds", y = "residuals",kind = "scatter")
```

Com base na Figura acima, não possível identificar nenhum padrão nas predições.

### Lasso

```python
model_lasso = LassoCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(X_train, y)
```

```python
rmse_cv(model_lasso).mean()
```

Como podemos observar, o modelo Lasso apresentou RMSE um pouco maior se comparado ao Ridge. 

```python
coef = pd.Series(model_lasso.coef_, index = X_train.columns)
```

```python
print("O modelo Lasso selecionou " + str(sum(coef != 0)) + " variáveis e eliminou outras" +  str(sum(coef == 0)) + " variáveis")
```

Vamos agora verificar os coefiecientes mais importântes para o modelo:

```python
imp_coef = pd.concat([coef.sort_values().head(10),
                     coef.sort_values().tail(10)])
```

```python
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Coefficients in the Lasso Model")
```

De acordo com a Figura apresentada acima, é possível perceber que algumas das variáveis com maior importância para o modelo são: media_despesa, quantidade_fornecedores, quantidade_doadores. Além disso, podemos observar, que algumas variáveis categorias apresentaram valores negativos, demostrando pouca/nenhuma importância para o modelo.

Vamos agora verifiar os resíduos obtidos pelo modelo.

```python
#let's look at the residuals as well:
matplotlib.rcParams['figure.figsize'] = (12.0, 6.0)

preds = pd.DataFrame({"preds":model_lasso.predict(X_train), "true":y})
preds["residuals"] = preds["true"] - preds["preds"]
preds.plot(x = "preds", y = "residuals",kind = "scatter")
```

### Dados de 2010 (Teste)

```python
X_test = test.drop('votos', axis=1)
```

```python
#Ridge
preds_ridge = model_ridge.predict(X_test)

# #Lassso
preds_lasso = model_lasso.predict(X_test)

# #Regressão simples
preds_linear_regression = model_linear_regression.predict(X_test)

# #KNN
preds_knn = knn_model.predict(X_test)

# #RANDOM FLOREST
preds_rn_florest = model_random_florest.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

ridge_rmse = mean_squared_error(test.votos, preds_ridge) ** 0.5
ridge_r2 = r2_score(test.votos, preds_ridge)

lasso_rmse = mean_squared_error(test.votos, preds_lasso) ** 0.5
lasso_r2 = r2_score(test.votos, preds_lasso)

linear_regression_rmse = mean_squared_error(test.votos, preds_linear_regression) ** 0.5
linear_regression_r2 = r2_score(test.votos, preds_linear_regression)

knn_rmse = mean_squared_error(test.votos, preds_knn) ** 0.5
knn_r2 = r2_score(test.votos, preds_knn)

knn_rmse = mean_squared_error(test.votos, preds_knn) ** 0.5
knn_r2 = r2_score(test.votos, preds_rn_florest)

print ("Modelo RIDGE: RMSE: ", ridge_rmse, "Rsquared: ", ridge_r2)
print ("Modelo LASSO: RMSE: ", lasso_rmse, "Rsquared: ", lasso_r2)
print ("Modelo REGRESSÃO LINEAR SIMPLES: RMSE: ", linear_regression_rmse, "Rsquared: ", linear_regression_r2)
print ("Modelo KNN: RMSE: ", knn_rmse, "Rsquared: ", knn_r2)
print ("Modelo RANDOM FLOREST: RMSE: ", knn_rmse, "Rsquared: ", knn_r2)
```

Como podemos observar o modelo com melhor resultado para o ano de 2010 foi Lasso com RMSE de 1.809
